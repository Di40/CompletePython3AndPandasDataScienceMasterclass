## Visualisations with Seaborn

```python
df2 = sns.load_dataset('iris')
df2.head()
```

```python
sns.displot(df2['petal_length'], kind="hist")
plt.show()
```

```python
sns.displot(df2['petal_length'], kind="kde")
plt.show()
```

```python
sns.displot(df2['petal_length'], kde=True)
plt.show()
```

## Plotting bivariate distribution

```python
sns.jointplot(x='petal_length', y='petal_width', data=df2)
plt.show()
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=df2, kind='hex')
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(df2, hue='species', diag_kind='kde', kind='scatter', palette='husl')
plt.show()
```

```python
plt.figure()
sns.stripplot(x='species', y='petal_length', data=df2)
plt.show()
```

```python
plt.figure()
sns.stripplot(x='species', y='petal_length', data=df2, jitter=True)
plt.show()
```

```python
plt.figure()
sns.swarmplot(x='species', y='petal_length', data=df2, size=3)
plt.show()
```

## Regression Visualized

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, row='sex')
plt.show()
```
