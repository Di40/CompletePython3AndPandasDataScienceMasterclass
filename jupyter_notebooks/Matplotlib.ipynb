# Matplotlib

```python
# Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
print(x)

y = x
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
N = 11
x_linear_range = np.linspace(0, 10, N, dtype=int)
print(x_linear_range)

y_linear_range = x_linear_range
plt.plot(x_linear_range, y_linear_range, 'o--')
plt.axis('off')
plt.show()
```

```python
y_logspace = np.logspace(0.1, 1, N)
print(y_logspace)
plt.plot(x_linear_range, y_logspace, 'o--')
plt.show()
```

```python
y_geomeric = np.geomspace(0.1, 1000, N)
print(y_geomeric)
plt.plot(x_linear_range, y_geomeric, 'ro--')
plt.show()
```

## Plotting in detail

```python
x = [1, 4, 5, 2, 3, 6]
plt.plot(x)
plt.show()
```

```python
x = np.arange(10)
plt.plot(x)
plt.show()

plt.plot(x,[y**2 for y in x])
plt.show()
```

```python
x = np.arange(-5,6)
plt.plot(x)
plt.show()

plt.plot(x,[y**2 for y in x]) # same as plt.plot(x, x**2)
plt.show()
```

## Multiline plots

```python
x = np.arange(10)
for i in range(2, 4):
    plt.plot(x, x**i)
for i in range(2, 4):
    plt.plot(x, x*i)
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2, x, x*3)
plt.show()
```

```python
x = np.array([[1, 2, 6], [4, 5, 3]])
plt.plot(x)
plt.show()
```

```python
data = np.random.randn(2, 10) # Random with normal distribution
print(data)
plt.plot(data[0], data[1])
plt.show()
```

```python
print(np.mean(data))
print(np.std(data))
print(np.var(data))
```

## Grid, axes and labels

```python
x = np.arange(3)
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.axis([0, 2, 0, 8])
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlim([0, 2])
plt.ylim([0, 8])
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.axis([0, 2, 0, 8])
plt.xlabel(' x = np.arange(3)')
plt.ylabel(' y = f(x)')
plt.title('Plot demo')
plt.show()
```

```python
plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.plot(x, 2*x, label = '2x')
plt.plot(x, 2**x, label = '2^x')
plt.legend()
plt.grid(True)
plt.axis([0, 2, 0, 8])
plt.xlabel(' x = np.arange(3)')
plt.ylabel(' y = f(x)')
plt.title('Plot demo')
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.legend(['x^2', 'x^3', '2x', '2^x'], loc = 'upper center') #positioning the legend
plt.grid(True)
plt.axis([0, 2, 0, 8])
plt.xlabel(' x = np.arange(3)')
plt.ylabel(' y = f(x)')
plt.title('Plot demo')
plt.savefig('matplotlib_pyplot_exercise') # save/export figure
plt.show()
```

## Color line markers

```python
x = np.arange(5)
y = x

# Can also be put in a single line
plt.plot(x, y + 1, 'g')
plt.plot(x, y + 0.5, 'y')
plt.plot(x, y, 'r')
plt.plot(x, y - 0.2, 'c')
plt.plot(x, y - 0.4, 'k')
plt.plot(x, y - 0.6, 'm')
plt.plot(x, y - 0.8, 'w') # white => not visible
plt.plot(x, y - 1, 'b')
plt.grid(True)
plt.show()
```

```python
y = x
for i in range(0, 10, 2): # range doesn't work with floats, so we cannot use (0, 1, 0.2)
    plt.plot(x, y + i/10)
    plt.plot(x, y - i/10)
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, '-', x, y + 1, '--', x, y + 2, '-.', x, y + 3, ':')
# single line, broken line, dash dot format, dot dot format
plt.grid(True)
plt.axis([0, 4, 0, 7])
plt.show()
```

```python
# All available marker types in matplotlib
plt.plot(x, y, '.')
plt.plot(x, y + 0.5, ',')
plt.plot(x, y + 1, 'o')
plt.plot(x, y + 2, '<')
plt.plot(x, y + 3, '>')
plt.plot(x, y + 4, 'v')
plt.plot(x, y + 5, '^')
plt.plot(x, y + 6, '1')
plt.plot(x, y + 7, '2')
plt.plot(x, y + 8, '3')
plt.plot(x, y + 9, '*')
plt.grid(True)
plt.show()
```

```python
# it can only hold these tree properties: line color, marker type, line style
plt.plot(x, y, 'mo--') # line color = magenta, marker type = 'o', line style = '--'
plt.plot(x, y + 0.5, 'g*-.') # line color = green, marker type = '*', line style = '-.'
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, color = 'g', linestyle = '--', linewidth = 1.5,
        marker = '^', markerfacecolor = 'w', markeredgecolor = 'k',
        markeredgewidth = 1.5, markersize = 5)
plt.grid(True)
plt.show()
```

```python
x = y = np.arange(10)
plt.plot(x, y, 'o--')
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
# list of tick locations, tick labels
plt.yticks(range(1, 10, 1))
plt.grid(True)
plt.show()
```
