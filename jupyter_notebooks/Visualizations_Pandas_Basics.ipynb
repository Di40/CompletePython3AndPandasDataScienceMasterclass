# Hexagonal Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000) # two randomizations (no specific reason for this)
plt.figure()
df.plot.hexbin(x='a', y='b', gridsize=10)
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df['z'] = np.random.uniform(0, 3, 1000) # colormap from 0 to 3
plt.figure()
df.plot.hexbin(x='a', y='b', C='z', reduce_C_function=np.max, gridsize=25)
plt.show()
```

# Pandas plotting

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
plt.figure()
scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde') # ‘kde’ or 'density' for density plots
# diagonal='hist'
plt.show()
```

```python
"""
Kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset,
analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.
"""
s1 = pd.Series(np.random.randn(1000))
plt.figure()
s1.plot.kde()
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('This is a somewhat long figure title', fontsize=16)
plt.subplot(1, 2, 1)
s1.plot.kde()
plt.subplot(1, 2, 2)
s1.plot.hist()
plt.show()
```

```python
plt.figure()
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))
lag_plot(data)
plt.show()
```

```python
plt.figure()
data = pd.Series(0.7 * np.random.rand(1000) + 0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))
autocorrelation_plot(data)
plt.show()
```

```python
data = pd.Series(np.random.rand(1000))
plt.figure()
bootstrap_plot(data, size=50, samples=500, color='purple')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
df = pd.DataFrame(np.random.rand(5, 3), columns=['a', 'b', 'c'])
ax.get_xaxis().set_visible(False)   # Hide Ticks on x axis
plt.figure()
df.plot(table=True, ax=ax)
plt.show()
```

```python
dd = pd.DataFrame(np.random.randn(10, 10)).applymap(abs) # applymap applies a function (absolute value) to a Dataframe elementwise.
dd = dd.cumsum()
plt.figure()
dd.plot.bar(colormap='Greens')
plt.show()
```

```python
price = pd.Series(np.random.randn(150).cumsum(),
                  index=pd.date_range('2000-1-1', periods=150, freq='B'))
# B - business day frequency
# https://pandas.pydata.org/docs/user_guide/timeseries.html#timeseries-offset-aliases
ma = price.rolling(20).mean()
mstd = price.rolling(20).std()
plt.figure()
plt.plot(price.index, price, 'k')
plt.plot(ma.index, ma, 'b')
plt.fill_between(mstd.index, ma-2*mstd, ma+2*mstd, color='b', alpha=0.2)
plt.show()
```
